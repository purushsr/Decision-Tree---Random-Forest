```python
import pandas as pd
import numpy as np
```

```python
dt = pd.read_csv('50_Startups.csv')
```

```python
dt.head()
```

```python
dt['State'] = dt['State'].astype('category')
dt['State'] = dt['State'].cat.codes
dt = pd.get_dummies(dt, columns=['State'])
```

```python
dt.head()
```

```python
dt = dt.drop(['State_0'],axis=1)
```

```python
dt.head()
```

```python
x = dt.drop(['Profit'],axis=1)
y = dt['Profit']
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=3)
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```
